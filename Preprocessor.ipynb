```python
import pandas as pd

# Load Movie Dataset
movies = pd.read_csv("./data/movies_metadata.csv", usecols=[5, 9, 20, 22, 23])#, nrows = 1000)

# Drop null values
movies = movies.dropna()

# Drop rows where overview is "No overview found." or "No overview"
movies = movies[(movies["overview"] != "No overview found.") & (movies["overview"] != "No overview")]

# Keep only movies with more than 10 votes (50% interval) and vote_average above 6 (above mean of 5.65 and at median of 6.0)
movies = movies[(movies["vote_count"] > 10) & (movies["vote_average"] > 1)]

# Remove duplicate titles, keep only the best score vote
mask = movies.duplicated(subset=["title"], keep=False)
movies = movies[~(mask & (movies["vote_average"] == movies[mask]["vote_average"].min()))]

# Assign indexes
movies["index"] = [i for i in range(0, len(movies))]

# Save file
movies.to_csv("./data/movies_metadata_preprocessed.csv")

movies
```
